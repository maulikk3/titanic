```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as ppt
import seaborn as sns
```

loading the dataset


```python
df=pd.read_csv("/content/tested.csv")
df.head()
```

```python
df.tail(15)
```

```python
df.dtypes
```

dropping the irrelevant columns

```python
pd.unique(df['PassengerId'])
```

```python
df=df.drop(['PassengerId','Fare'],axis=1)
```

```python
df.head()
```

delete the duplicate rows

```python
df.shape
```

```python
duplicate_rows_df=df[df.duplicated()]
print("number of duplicate rows",duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df.drop_duplicates()
df.head()
```

```python
# droppping the missing  or null values
print(df.isnull().sum())
```

```python
df=df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```
